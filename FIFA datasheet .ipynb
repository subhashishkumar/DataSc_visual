### Graphical representation on the proportionality of Player's 'Age' and 'Overall Potential'

```python
rbao = fl.groupby('Age')['Overall'].mean().reset_index()
rbao.sort_values('Age')
plt.plot(rbao['Age'], rbao['Overall'])
plt.xlabel('Age Group')
plt.ylabel('potential')
plt.show()
```

# Relation between AGE and POTENTIAL

```python
rbap = fl.groupby('Age')['Potential'].mean().reset_index()
rbap.sort_values('Age')
plt.plot(rbap['Age'], rbap['Potential'])
plt.show()
```
